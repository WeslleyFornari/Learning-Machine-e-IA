```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# X_credit_treinamento, X_credit_teste, y_credit_treinamento, y_credit_teste = train_test_split
# (X_credit, y_credit, test_size = 0.25, random_state = 0)

# Deserializa os dados binarios para os arrays predefinidos
import pickle
with open('base_dados/credit.pkl', 'rb') as h:
  X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(h)
```

```python
# X - pre processamento (quant, colunas) Y- classes -> SHAPE

X_credit_treinamento.shape, y_credit_treinamento.shape
```

```python
# Testes
X_credit_teste.shape, y_credit_teste.shape
```

```python
# Library de Learning MAchine / algoritmo
from sklearn.naive_bayes import GaussianNB
```

```python
# Algoritmo de treinamento / Aprendizado   metodo FIT
naive_credit_data = GaussianNB()
naive_credit_data.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
# predict( ... ) aplica na base de dados 
previsoes = naive_credit_data.predict(X_credit_teste)
```

```python
# 1 - não paga     2 - paga
previsoes
```

```python
# Comparar com os resultados acima - Previsão X Respostas
y_credit_teste
```

```python
# Medições e Metricas
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Compara - (Respostas, Previsões) - % de acertos 93,8 % de 500 registros
accuracy_score(y_credit_teste, previsoes)
```

```python
confusion_matrix(y_credit_teste, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
cm = ConfusionMatrix(naive_credit_data) # Algoritimo
cm.fit(X_credit_treinamento, y_credit_treinamento) # Treinamento
cm.score(X_credit_teste, y_credit_teste) # resultado

# 93,8% acerto. 0- Paga  1-Não Paga    Score Antigo -> Classificação Atual 
```

```python
# Relatorios planilha
print(classification_report(y_credit_teste, previsoes))
```
